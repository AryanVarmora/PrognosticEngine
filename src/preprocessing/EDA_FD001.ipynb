```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the preprocessed training data
train_data = pd.read_csv("/Users/aryan/Desktop/PrognosticEngine/data/train_FD001_preprocessed.csv")

# Basic info and preview of the dataset
print("Dataset Information:")
print(train_data.info())

print("\nPreview of the Dataset:")
print(train_data.head())

# Plot Distribution of Remaining Useful Life (RUL)
plt.figure(figsize=(10, 6))
sns.histplot(train_data['RUL'], kde=True, bins=30, color='blue')
plt.title('Distribution of Remaining Useful Life (RUL)', fontsize=16)
plt.xlabel('RUL', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Correlation Matrix for Sensor Measurements
sensor_columns = [col for col in train_data.columns if 'sensor_measurement' in col]
correlation_matrix = train_data[sensor_columns + ['RUL']].corr()

# Heatmap for Correlations
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap: Sensors and RUL', fontsize=16)
plt.show()
```
